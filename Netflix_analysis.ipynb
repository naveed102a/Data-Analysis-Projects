#### Q: in which year highest number of TV shows & Movies were relased ? show with bar graph

- dtype

```python
data.dtypes
```

```python
#here we create a new column for date time format
#errors='coerce' hmny islye use kiya hy column mein kuch values ka format same nahi hai ya unke start mein extra space hai. 
#taake jo date convert na ho wo NaT (missing) ban jaye aur error na aaye.

data['Date_N'] = pd.to_datetime(data['Release_Date'], errors='coerce') 
```

```python
data.head(3)
```

- dt.year.value_counts()

```python
data['Date_N'].dt.year.value_counts()  # it counts the accourance of all individual yearsin date column 
```

```python
plt.figure(figsize=(6,4))
data['Date_N'].dt.year.value_counts().plot(kind = 'bar')
```
